#### readline()

```python
f = open('./Ch1_2.ipynb', mode='rt', encoding='utf-8')
```

```python
while True:
    line = f.readline()
    if not line:
        break;
    line = line.strip()
    print(line)
    
    
f.close()
```

#### append()

```python
f = open('./test.txt', mode='a', encoding='utf-8')
```

```python
for i in range(10, 50):
    line = f'{i} 번째 줄입니다\n'
    f.write(line)
f.close()
```
